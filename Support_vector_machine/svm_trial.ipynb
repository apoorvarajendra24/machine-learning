```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create synthetic data
from sklearn.datasets import make_classification

# Generate linearly separable data for SVM
X, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    n_classes=2,
    class_sep=2.0,
    #flip_y=0,
    random_state=4
)

# Convert to DataFrame for easier plotting
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['Target'] = y

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Feature1', y='Feature2', hue='Target', data=df, palette='coolwarm', edgecolor='k')
plt.title("Linearly Separable Data for SVM")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
# Train test split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# SVM(SVC) Model
from sklearn.svm import SVC
svc = SVC(kernel="linear")
```

```python
# Train the model
svc.fit(X_train, y_train)
```

```python
# Predict the values
y_pred = svc.predict(X_test)
print("Predicted values: ", y_pred)
```

```python
# Performance matrices
from sklearn.metrics import classification_report, confusion_matrix
print(f'Confusion Matrix : \n{confusion_matrix(y_test,y_pred)}')
print(f'Classification Report : {classification_report(y_test,y_pred)}')

# Since the data is already so categorised the accuracy is solid.
```

```python
X, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=2,
    n_classes=2,
    class_sep=2.0,
    #flip_y=0,
    random_state=4
)

# Convert to DataFrame for easier plotting
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['Target'] = y

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Feature1', y='Feature2', hue='Target', data=df, palette='coolwarm', edgecolor='k')
plt.title("Linearly Separable Data for SVM")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# SVC linear kernel Model
linear_ker = SVC(kernel="linear")
linear_ker.fit(X_train, y_train)
y_pred1 = linear_ker.predict(X_test)
```

```python
print(f'Predicted values : {y_pred1}')

# Performance matrices
print(f'Confusion Matrix : \n{confusion_matrix(y_test,y_pred1)}')
print(f'Classification Report : {classification_report(y_test,y_pred1)}')
```

```python
# SVC Polynomial kernel model
poly_ker = SVC(kernel="poly")
poly_ker.fit(X_train,y_train)
y_pred2 = poly_ker.predict(X_test)
```

```python
print(f'Predicted values : {y_pred2}')

# Performance matrices
print(f'Confusion Matrix : \n{confusion_matrix(y_test,y_pred2)}')
print(f'Classification Report : {classification_report(y_test,y_pred2)}')
```

```python
# SVC rbf kernel model
rbf_ker = SVC(kernel="rbf")
rbf_ker.fit(X_train,y_train)
y_pred3 = rbf_ker.predict(X_test)
```

```python
print(f'Predicted values : {y_pred3}')

# Performance matrices
print(f'Confusion Matrix : \n{confusion_matrix(y_test,y_pred3)}')
print(f'Classification Report : {classification_report(y_test,y_pred3)}')
```

```python
# SVC Sigmoid kernel model
sig_ker = SVC(kernel="sigmoid")
sig_ker.fit(X_train,y_train)
y_pred4 = sig_ker.predict(X_test)
```

```python
print(f'Predicted values : {y_pred4}')

# Performance matrices
print(f'Confusion Matrix : \n{confusion_matrix(y_test,y_pred4)}')
print(f'Classification Report : {classification_report(y_test,y_pred4)}')
```
