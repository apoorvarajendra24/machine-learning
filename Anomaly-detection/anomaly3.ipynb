```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
X,y = make_circles(n_samples=300,factor=0.5,noise=0.05,random_state=42)
X
```

```python
np.random.seed(42)
outliers = np.random.uniform(low=-1.5, high=1.5, size=(20,2))
X = np.concatenate([X,outliers], axis=0)
X
```

```python
dbscan = DBSCAN(eps=0.15, n_jobs=-1, min_samples=5)
dbscan.fit(X)
```

```python
labels = dbscan.labels_
labels
```

```python
anomalies = X[labels == -1]
normal = X[labels != -1]

plt.figure(figsize=(8,6))
plt.scatter(normal[:,0],normal[:,1], c="b",s=40, label="Normal")
plt.scatter(anomalies[:,0],anomalies[:,1],c="r",s=80,label="Anomaly",marker='X')
plt.title("Anomaly detetion (DBSCAN)")
plt.legend()
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.grid(True)
plt.show()
```

```python
print(f"Number of anomalies: {len(anomalies)}")
```
