```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set seed for reproducibility
np.random.seed(42)
```

```python
# Generate 200 normal data points around (2,2) and (-2,-2)
normal_data = 0.5 * np.random.randn(100, 2) + [2, 2]
normal_data = np.vstack([normal_data, 0.5 * np.random.randn(100, 2) + [-2, -2]])
```

```python
# Generate 20 anomaly data points spread out
anomaly_data = np.random.uniform(low=-6, high=6, size=(20, 2))
```

```python
# Combine them
X = np.vstack([normal_data, anomaly_data])
y = np.array([1]*200 + [-1]*20)  # 1 = normal, -1 = anomaly
```

```python
# Plot
plt.scatter(X[y==1][:, 0], X[y==1][:, 1], c='blue', label='Normal')
plt.scatter(X[y==-1][:, 0], X[y==-1][:, 1], c='red', label='Anomaly')
plt.title("Synthetic Dataset with Anomalies")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(contamination=0.2,random_state=42)  # Anything above 0.2 is considered as an anomaly
clf.fit(X)
prediction = clf.predict(X)
```

```python
prediction
```

```python
import numpy as np
abn_ind = np.where(prediction<0)
abn_ind
```

```python
plt.scatter(X[:,0],X[:,1],c='blue', label='normal')
plt.scatter(X[abn_ind,0],X[abn_ind,1],c='red',label='anomaly')
plt.legend()
plt.grid(True)
plt.show()
```
