```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score
import seaborn as sns 
```

```python
milk = pd.read_csv("milk.csv",index_col=0)
milk.head()
```

```python
scaler = StandardScaler().set_output(transform='pandas')
scaler.fit(milk)
milk_scaled=scaler.transform(milk)
```

```python
silhoutees_coefficient = []
ks = [2,3,4,5,6,7,8,9,10]
for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=24)
    kmeans.fit(milk_scaled)
    silhoutees_coefficient.append(silhouette_score(milk_scaled, kmeans.labels_))

print(silhoutees_coefficient)
print(kmeans.cluster_centers_)
print(kmeans.labels_)
```

```python
k_max = np.argmax(silhoutees_coefficient)
print(f"Best Score : {silhoutees_coefficient[k_max]}")
print(f'Best value of K : {ks[k_max]}')
```

```python
kmeans = KMeans(n_clusters=2,init="k-means++" ,random_state=24)
kmeans.fit(milk_scaled)
```

```python
kmeans_data = milk.copy()
kmeans_data['Clust'] = kmeans.labels_
print( kmeans_data.groupby('Clust').mean() )
```
